## 深度学习简介

目前， 在计算机编程解决实际问题领域内，编程方式有两类：

- 基于“常识”编程
- 基于“数据”编程

下面我们来详细看看这两类编程方式的差别：

**基于常识编程**

基于常识编程是目前最常见的一种编程方式。目前，市面上各大培训机构培养的程序员，都是能熟练掌握这种编程方式的人员。这一类编程方式，只要你具有业务的基本逻辑性，一般都能使用这种编程方式解决问题，比如说，我们要开发一个邮件系统，邮件系统的实现逻辑如下：

- 如何输入收件人地址，邮件的标题、邮件的正文
- 如何触发发送邮件
- 如何检查用户的输入是否正确
- 如何发送邮件到收件人的地址

这些问题很明确的，只要编程人员有充足的时间、足够好的编程技巧、足够多的常识，这些问题解决起来是非常容易的。

**基于数据编程**

基于数据编程是利用大量数据之间存在某种逻辑关系的思想进行解决问题的编程方式。比如说，识别图像中动物是猫还是狗，这就是要找出图像中的像素之间是否存在某种关系，而这种关系能用来表示图像中的内容是猫还是狗。

解决这一类问题，需要编程人员有很高的逻辑性和丰富的统计学知识。因此，无论你有多么好的编程技巧，有多么充裕的时间，对于一个没有基于数据编程经验的人来说，都是不可能解决这类问题的。

在解决这类问题的过程中，需要使用到大量的数据，并通过大量的数学知识抽象出数据间的关系，而这个过程中，需要用到的算力，远远超出了人类的计算上限。因此，我们借助工具--计算机来完成这类问题的解答，因此，基于数据编程的方式也被称为机器学习。


### 起源

基于数据编程的核心思想已经被人类研究了数百年。自古以来，人们就一直渴望能从数据中分析出预知未来的窍门。实际上，数据分析正是大部分自然科学的本质，人们希望从日常的观测中提取规则，并找寻不确定性

**最小二乘法** - 线性回归的鼻祖

1801年，意大利天文学家朱赛普·皮亚齐发现了第一颗小行星--谷神星。经过40天的跟踪观测后，由于谷神星运行至太阳背后，使得皮亚齐失去了谷神星的位置。随后全世界的科学家利用皮亚齐的观测数据开始寻找谷神星，但是根据大多数人计算的结果来寻找谷神星都没有结果。只有时年24岁的高斯所计算的谷神星的轨道，被奥地利天文学家海因里希·奥尔伯斯的观测所证实，使天文界从此可以预测到谷神星的精确位置。同样的方法也产生了哈雷彗星等很多天文学成果。高斯使用的方法就是最小二乘法，该方法发表于1809年他的著作《天体运动论》中 。1829年，高斯提供了最小二乘法的优化效果强于其他方法的证明 。
